```python
import requests
from bs4 import BeautifulSoup
from textblob import TextBlob

arr=['https://twitter.com/ETMarkets','https://twitter.com/TimesNow?lang=en','https://twitter.com/BloombergQuint?lang=en']
for k in arr:    
    r= requests.get(k)
    soup= BeautifulSoup(r.text, 'html.parser')
    print(soup.title)
    i=1
    for tweets in soup.findAll('div',{'class':'content'}):
        if 'market' in tweets.find('p').text:
            print(i)
            print(tweets.find('p').text)
            text=TextBlob(tweets.find('p').text)
            #print(text.words)
            print(text.sentiment.polarity)
            i+=1
```
